## Filter the Accidents in Georgia

```python
import pandas as pd
import json
```

```python
def load_data_pd(path):
    # read data using pandas
    # df = pd.read_csv(path, nrows=1000)
    df = pd.read_csv(path)
    print(df['State'].unique())

    return df.loc[df['State'] == 'GA']
```

```python
df = load_data_pd("dataset/US_Accidents_March23.csv")
display(df)
```

```python
df.to_csv("dataset/GA_Accidents_March23.csv", index=False)
```

## Create a Joint JSON File of Edges and Accidents

```python
# save nodes to a json file
df = pd.read_csv("osm/nodes.csv")
# print(df)
# df.to_json("osm/nodes.json", orient='')
id_list = df['id'].tolist()
ln_list = df['lon'].tolist()
la_list = df['lat'].tolist()
print(len(id_list), len(ln_list), len(la_list))
```

```python
print(df['id'][0])
data = {}

for id, lon, lat in zip(id_list, ln_list, la_list):
    data[id] = {'lon': lon, 'lat': lat}

with open("osm/nodes.json", "w") as json_file:
    json.dump(data, json_file, indent=4)
```

```python
# read edges file
df_edg = pd.read_csv("osm/edges.csv")

id_list = df_edg['id'].tolist()
src_list = df_edg['source'].tolist()
tar_list = df_edg['target'].tolist()

# read nodes file
with open ("osm/nodes.json", "r") as f:
    nodes = json.load(f)
```

```python
data = {}
for id, src, tar in zip(id_list, src_list, tar_list):
    edge = {}
    x1, y1 = float(nodes[str(src)]['lon']), float(nodes[str(src)]['lat'])
    x2, y2 = float(nodes[str(tar)]['lon']), float(nodes[str(tar)]['lat'])
    edge['src'] = (x1, y1)
    edge['tar'] = (x2, y2)
    # slope and intercept
    # a, b, c = calculate_line_parameters(x1, y1, x2, y2)
    # edge['line'] = (a, b, c)
    edge['acc'] = []
    data[id] = edge

with open("osm/edges.json", "w") as f:
    json.dump(data, f, indent=4)
```
